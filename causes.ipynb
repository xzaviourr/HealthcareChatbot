```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
import string
```

```python
disease_df = pd.read_json('Data/disease_dataset_full.json')
disease_df.head()
```

```python
causes = disease_df['cause']
causes.head()
```

```python
def preprocess_causes(cause):
    if cause == None:
        cause = "No information"
    return cause

causes = causes.apply(preprocess_causes)
```

```python
lemmer = nltk.stem.WordNetLemmatizer()
def LemTokens(word_tokens):
    return [lemmer.lemmatize(token) for token in word_tokens]
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)

def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))


TfidfVec = TfidfVectorizer(tokenizer=LemNormalize, stop_words='english')
tfidf = TfidfVec.fit_transform(causes) 
```

```python
import numpy as np

labels = TfidfVec.get_feature_names()
labels = np.array(labels)

i = 0
for row in tfidf:
    filter = row.toarray()
    indexes = np.where(filter > 0.1)[1]
    indexes = indexes.astype(int)
    if indexes != []:
        print(causes[i])
        print(labels[indexes])
    i += 1

    break
```
