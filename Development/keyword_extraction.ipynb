```python
import spacy
import pandas as pd     
```

```python
disease_df = pd.read_json('Data/disease_dataset_full.json')
disease_df.head()
```

```python
causes = disease_df['cause']
causes.head()
```

```python
x = list(causes[:1])
x
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(x[0]).ents
doc
```

```python
nlp = spacy.load("en_core_web_sm")
def extract_keywords(text):
    doc = nlp(text).ents
    return doc

keywords = causes.apply(extract_keywords)
keywords
```
