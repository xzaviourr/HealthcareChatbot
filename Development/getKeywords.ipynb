```python
import pandas as pd
disease_df = pd.read_json('../Data/disease_dataset_full.json')
disease_df.head()
```

```python
import yake

def yake_extractor(text):
    if text==None:
        return []
    kw_extractor = yake.KeywordExtractor()
    #text = """spaCy is an open-source software library for advanced natural language processing, 
    # written in the programming languages Python and Cython. 
    # The library is published under the MIT license and its main developers are Matthew Honnibal and Ines Montani, 
    # the founders of the software company Explosion."""
    language = "en"
    max_ngram_size = 3
    deduplication_threshold = 0.9
    numOfKeywords = 20
    custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
    res=custom_kw_extractor.extract_keywords(text)
    final_lst=[]
    for r in res:
        final_lst.append(r[0])
    return final_lst

def cause_preprocess(text):
    if text==None:
        return text
    return text.replace('\n', ' ')

```

```python
disease_df['cause']= disease_df['cause'].apply(cause_preprocess)
```

```python
disease_df['cost_money']= disease_df['cost_money'].str.split('[()]').str[1].str.findall('(\d+)')
disease_df['get_prob']= disease_df['get_prob'].apply(lambda x: x[:-1] if x!=None else x)
```

```python
disease_df['prevent'] = disease_df['prevent'].apply(yake_extractor)
disease_df['cause']= disease_df['cause'].apply(yake_extractor)
disease_df['easy_get'] = disease_df['easy_get'].apply(yake_extractor)
```
